```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv("tracks.csv")
df_tracks.head()
```

```python
# check for null values
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
# finding 10 least popular songs using sort_values() function
sort_df = df_tracks.sort_values('popularity').head(10)
sort_df['name']
```

```python
df_tracks.describe().T
```

```python
# top 10 most popular songs (popularity > 90)

popular_df = df_tracks.query('popularity > 90', inplace=False).sort_values('popularity', ascending=False)
popular_df.head(10)
```

```python
# set release date as index

df_tracks.set_index("release_date", inplace=True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks.index
```

```python
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms', inplace=True, axis=1)
df_tracks.head()
```

```python
df_corr = df_tracks.drop(['key', 'mode', 'explicit'], axis=1).corr(method='pearson')
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(df_corr, annot=True, vmin=-1, vmax=1, center=0,
                      cmap='inferno', linewidths=1, linecolor='Black')
heatmap.set_title('Correlation heatmap between variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90);
```

```python
# sampling subset of the original dataset

df_sample = df_tracks.sample(int(0.004*len(df_tracks)))
print(len(df_sample))
```

```python
# regression line plot between loudness and energy

plt.figure(figsize=(10, 6))
sns.regplot(data=df_sample, x='energy', y='loudness', color='c').set(title='Loudness vs Energy regression');
```

```python
# regression line plot between acousticness and popularity

plt.figure(figsize=(10, 6))
sns.regplot(data=df_sample, x='acousticness', y='popularity',
            color='c').set(title='acousticness vs popularity regression');
```

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
```

```python
df_tracks.dates
```

```python
years = df_tracks.dates.dt.year
```

```python
# number of songs released over the years

sns.displot(years, discrete=True, aspect=2, height=5, kind='hist').set(title='Number of songs per year');
```

```python
# duration of the songs over the years

total_dr = df_tracks.duration
sns.set_style(style='whitegrid')
fig, ax = plt.subplots(figsize=(18, 7))
fig = sns.barplot(x=years, y=total_dr, ax=ax, errwidth=False).set(title='Years vs Duration')
plt.xticks(rotation=90);
```

```python
df_genre = pd.read_csv('SpotifyFeatures.csv')
```

```python
df_genre.head()
```

```python
# duration of songs for different genre

plt.figure(figsize=(12, 6))
sns.color_palette('rocket', as_cmap=True)
sns.barplot(data=df_genre, x='duration_ms', y='genre')
plt.title('Duration of songs for different genre')
plt.xlabel('Duration in ms')
plt.ylabel('Genre');
```

```python
# top 5 genre by popularity

sns.set_style(style='darkgrid')
plt.figure(figsize=(12, 6))
famous = df_genre.sort_values('popularity', ascending=False).head(10)
sns.barplot(y='genre', x='popularity', data=famous);
```
